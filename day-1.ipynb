# **Task 1** – Identify and handle missing values using .isnull() in Python or filters in Excel.

```python
#identifying missing 
print(df1.isnull().sum())
null_columns = df1.columns[df1.isnull().any()]
print("Columns with nulls:", null_columns.tolist())
```

```python
#option 1 – dropping missing values
df1_cleaned = df1.dropna()
print(df1_cleaned.info())
print(df1_cleaned.describe())
print(df1_cleaned.isnull().sum())

```

```python
#option 2 – filling missing values with a specific value 
#adding string values for director, cast, country date_added, rating, duration AS "Unknown"
df1['director'] = df1['director'].fillna('Unknown')
df1['cast'] = df1['cast'].fillna('Unknown')
df1['country'] = df1['country'].fillna('Unknown')
df1['date_added'] = df1['date_added'].fillna('Unknown')
df1['rating'] = df1['rating'].fillna('Unknown')
df1['duration'] = df1['duration'].fillna('Unknown')
```

```python
#checking for null values
print(df1.isnull().sum())
```

# **Task 2** - Remove duplicate rows using .drop_duplicates() or Excel’s“Remove Duplicates”.

```python

df1.drop_duplicates(inplace=True)
print(df1.info())
```
